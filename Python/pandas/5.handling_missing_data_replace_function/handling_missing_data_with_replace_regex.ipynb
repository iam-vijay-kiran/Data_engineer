```python
import pandas as pd
import numpy as np
df = pd.read_csv('weather_data.csv')
df
```

```python
# replacing multiple values using list & nan using numpy

new_df = df.replace([-99999,-88888],np.nan)
new_df
```

```python
# replace multiple values of multiple columns by specifying them in a dictionary

new_df = df.replace({"temperature":-99999,
                    "windspeed": [-88888,-99999],
                    "event":'0'},np.nan)
new_df
```

```python
# specifying required cell value to replace with required value and in that dict the key cannot be taken as list so we specify every value
new_df = df.replace({-99999:np.nan,-88888:np.nan,
                    '0':'Sunny'})
new_df
```
