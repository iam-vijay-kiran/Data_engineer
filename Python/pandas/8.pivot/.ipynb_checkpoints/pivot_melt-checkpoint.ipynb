# Pivot basics - changing dimension of table like row to column or vice versa

```python
import pandas as pd

df = pd.read_csv("weather.csv")
df
```

## table changes with entire data

```python
df.pivot(index="city", columns = "date")
```

## only humidity values are taken

```python
df.pivot(index="city",columns="date",values="humidity")
```

```python
df.pivot(index="date",columns="city")
```

```python
df.pivot(index="humidity",columns="city")
```

## pivot table 

```python
df = pd.read_csv("weather2.csv")
df
```

## pivot_table gives mean(average value) by default

```python
df.pivot_table(index="city",columns="date")
```

## we can explicitly control with aggfunc which value to calculate in pivot_table

```python
df.pivot_table(index="city",columns="date",aggfunc="sum")
```

## Margins - give column and row level entire aggfunc

```python
df.pivot_table(index="city",columns="date",margins=True,aggfunc="sum")
```

## Grouper - groups specified columns to aggfunc calculation by default it gives mean

```python
df = pd.read_csv("weather3.csv")
df
```

```python
# convert date to datetime format
df['date'] = pd.to_datetime(df['date'])
type(df['date'][0])
```

```python
## groups data based on date and finds thier mean

df.pivot_table(index=pd.Grouper(freq='ME',key="date"),columns='city')
```
