```python
# changed day column from string type to timestamp(date) type
import pandas as pd
df = pd.read_csv("weather_data.csv",parse_dates=["day"])
print(type(df.day[0]))
print(df)
```

```python
# replacing day as index
df.set_index('day',inplace=True)
df
```

## fillna

```python
# replacing all not available data with 0
new_df = df.fillna(0)
new_df
```

```python
# replacing specific na data with specific values

new_df = df.fillna({
    'temperature': 0,
    'windspeed': 0,
    'event':'no_event'
})
new_df
```

```python
# forward filling = filling empty cells with next cells data in entire dataframe
# new_df = df.fillna(method='ffill')
new_df = df.ffill()
new_df
```

```python
# backfilling similar to back
new_df = df.bfill()
new_df
```

```python
# axis is either "index" or "columns"
new_df = df.fillna(method="bfill", axis="columns")
new_df
```

```python
# copy only to next 1 row
new_df = df.fillna(method="ffill",limit=1)
new_df
```

## Interpolate

```python
new_df = df.interpolate()
new_df
```

```python
new_df = df.interpolate(method="time")
new_df
```

### dropna

```python
## drops atleast if it finds one null value
new_df = df.dropna()
new_df
```

```python
## drops the row where all columns value are missing
new_df = df.dropna(how="all")
new_df
```

```python
## keeps rows which has atleast 1 column value in it
new_df = df.dropna(thresh=1)
new_df
```
