# Concat

```python
## creating dataframes

import pandas as pd

india_weather = pd.DataFrame({
    "city": ["mumbai","delhi","banglore"],
    "temperature": [32,45,30],
    "humidity": [80, 60, 78]
})
india_weather
```

```python
us_weather = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
    "humidity": [68, 65, 75]
})
us_weather
```

## concat but here the indexing is seperated as per 2 df

```python
df = pd.concat([india_weather,us_weather])
df
```

## ignore index - we get continuous index

```python
df = pd.concat([india_weather,us_weather],ignore_index=True)
df
```

## Concatinating with keys

```python
df = pd.concat([india_weather,us_weather],keys=['india','usa'])
df
```

```python
df.loc['india']
```

## Concatination using index

```python
temperature_df = pd.DataFrame({
    "city": ["mumbai","delhi","banglore"],
    "temperature": [32,45,30],
}, index = [0,1,2])
temperature_df
```

```python
windspeed_df = pd.DataFrame({
    "city": ["delhi","mumbai"],
    "windspeed": [7,12],
},index = [1,0])
windspeed_df
```

```python
# axis = 1 says to append the new df as column for existing df else it appends data as new rows
df = pd.concat([temperature_df,windspeed_df],axis=1)
df
```

## concatination dataframe with series

```python
series = pd.Series(["Humid","Dry","Rain"], name="event")
series
```

```python
# here appends data as new rows
df = pd.concat([temperature_df,series])
df
```

```python
# appending as new column
df = pd.concat([temperature_df,series],axis=1)
df
```

# Merge - kind of sql joins

```python
import pandas as pd
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","orlando"],
    "humidity": [65,68,75],
})
df2
```

```python
df3 = pd.merge(df1,df2,on="city")  # this gives only common columns which is nothing but an inner join by default
df3
```

## database joins 

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35, 38],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "humidity": [65,68,71],
})
df2
```

```python
# inner join
df3 = pd.merge(df1,df2,on="city",how="inner")
df3
```

```python
# outer join
df4 = pd.merge(df1,df2, on="city",how="outer")
df4
```

```python
# left join
df5 = pd.merge(df1,df2, on="city",how="left")
df5
```

```python
# right join
df6 = pd.merge(df1,df2, on="city",how="right")
df6
```

## Indicator Flag - indicates the data origin either from left or right df

```python
df4 = pd.merge(df1,df2, on="city",how="outer",indicator=True)
df4
```

# Suffixes

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35,38],
    "humidity": [65,68,71, 75]
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "temperature": [21,14,35],
    "humidity": [65,68,71]
})
df2
```

```python
df3 = pd.merge(df1,df2, on="city",how="outer",suffixes=('_first','_second'))
df3
```
