# Joins - Inner, Outer, Left, Right

```python
df1 = pd.DataFrame({
    'id':[1,2,3,4],
    'name':['vijay','kiran','teja','rusum']
})

df2 = pd.DataFrame({
    'id': [3,4,5,6],
    'name': ['ca','ai','python','comm']
})

inner_result = pd.merge(df1,df2,how='inner',on='id')
inner_result
```

```python
outer_join = pd.merge(df1,df2,how='outer',on='id')
outer_join
```

```python
right_join = pd.merge(df1,df2,how='right',on='id')
right_join
```

```python
left_join = pd.merge(df1,df2,how='left',on='id')
left_join
```

## join on different id names

```python
df3 = pd.DataFrame({
    'id1':[1,2,3,4],
    'name':['vijay','kiran','teja','rusum']
})

df4 = pd.DataFrame({
    'id2': [3,4,5,6],
    'name': ['ca','ai','python','comm']
})

inner_result = pd.merge(df3,df4,how='inner',left_on='id1',right_on='id2')
inner_result
```

# Concatinating 2 dataframes

```python
df1 = pd.DataFrame({
    'id':[1,2,3,4],
    'name':['vijay','kiran','teja','rusum']
})

df2 = pd.DataFrame({
    'id': [3,4,5,6],
    'name': ['ca','ai','python','comm']
})

# axis 0,1 are stack one on another, side by side respectively
# ignore_index = gives continuous indexing else it gives indicies of respective df

result_1 = pd.concat([df1,df2],axis=0,ignore_index=True)
result_1
```

```python
result_2 = pd.concat([df1,df2],axis=1)
result_2
```

```python
# creates multilevel indexing like partitioin based on keys specified
result_3 = pd.concat([df1,df2],axis=0,keys=['df1','df2'])
result_3
```
