```python
'''
https://leetcode.com/problems/number-of-students-unable-to-eat-lunch/
https://www.youtube.com/watch?v=d_cvtFwnOZg
T - O(1) as we use queues, Space - O(1) as we are using only 2 keys as counter
'''

class Solution:
    def countStudents(self, students: List[int], sandwiches: List[int]) -> int:
        res = len(students)             # counting no.of. students
        cnt = Counter(students)         # creates dict by counting different individual variables

        # that counter through if loop
        # cnt = {}
        # for s in students:
        #     if s not in cnt:
        #         cnt[s] = 0
        #     cnt[s] += 1

        for s in sandwiches:
            if cnt[s] > 0:          # check whether that s available in cnt dict
                res -= 1            # if available student decrements
                cnt[s] -= 1         # cnt of that sandwich also decrements like dedqueue
            else:
                return res          # if not so stop & return the cnt of students i.e., res
        return res                  # show them res in output


'''
https://leetcode.com/problems/implement-stack-using-queues/
https://www.youtube.com/watch?v=rW4vm0-DLYc
T - O(n) as pop takes higher time complexity rest are O(1), S- O(1),where we dont use extra space
'''

class MyStack:

    def __init__(self):
        self.q = deque()        # initializing deque

    def push(self, x: int) -> None:
        self.q.append(x)            # for push we append it

    def pop(self) -> int:
        for i in range(len(self.q) -1 ):    # we just take range till last but 1 value i.e., [-2]
            self.push(self.q.popleft())     # using popleft we pop the element from left i.,e starting & push it from right to left like queue
        return self.q.popleft()             # later all the elements pushes from right of queue the starting is element to pop
                                            # which we pop using popleft. this takes T- O(n)

    def top(self) -> int:
        return self.q[-1]           # we can get last element -1 in python for top

    def empty(self) -> bool:
        return len(self.q) == 0     # check length of queue is empty & reutrn bool value


# Your MyStack object will be instantiated and called as such:
# obj = MyStack()
# obj.push(x)
# param_2 = obj.pop()
# param_3 = obj.top()
# param_4 = obj.empty()
```
