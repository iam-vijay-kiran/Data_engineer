# Arrays and their time complexity

### 1. Collection of similar datatypes at consicutive memory location is called an array
### 2. Python has only Dynamic arrays which means the size of array is not pre-defined and flexible
+ Array with fixed memory and cannot add more elements without adjusting the existing elements
- Static arrays are used when memory is fixed and prior known
- Dynamic arrays are used when size of memory is unknown prior and are defalut in python & javascript
- Ambratization - doubling the array size to increase space capacity which is **O(N)** inserting there is **O(1)**

```python
''' Remove duplicates = https://leetcode.com/problems/remove-duplicates-from-sorted-array
    Time complexity = O(n) as it moves through each element
    Space complexity = O(1) as it make operations in existing array itself '''
nums=[1,2,2,3,4,4,5]
class Solution:
    def removeDuplicates(self, nums: List[int]) -> int:
        l = 1
        for r in range(1,len(nums)):
            if nums[r] != nums[r-1]:
                nums[l] = nums[r]
                l +=1
        return l

'''        
 we initiliaze l,r at 1 not 0 because whenever we update the 0th value it stays in same 
 position so, we are checking from 1st position.      
 taking l as left pointer and comparing with r as right pointer if they are not equal 
 placing the element in left pointer 
'''


'''
 https://leetcode.com/problems/remove-element/
 removes the val element from list
 Time complexity = O(n) as it moves through each element'''
nums = [0,1,2,2,3,0,4,2], val = 2
class Solution:
    def removeElement(self, nums: List[int], val: int) -> int:
        k = 0
        for i in range(len(nums)):
            if nums[i] != val:
                nums[k] = nums[i]
                k += 1
        return k

    
'''
 https://leetcode.com/problems/concatenation-of-array/submissions/1232002497/
 https://www.youtube.com/watch?v=68isPRHgcFQ
 Time complexity = O(n) as we iterate through each element for both below codes
'''
class Solution:
    def getConcatenation(self, nums: List[int]) -> List[int]:
        return nums * 2
         
          #or 
        
class Solution:
    def getConcatenation(self, nums: List[int]) -> List[int]:
        ans = []
        for i in range(2):
            for n in nums:
                ans.append(n)
        return ans
```

# Stacks & their time complexity

- In python stacks are implimented using arrays(lists) which are dynamic in nature.so space wont be a problem

# Finally The Main difference between Arrays and Stacks with list in python

+ The main difference lines in the implementation perspective. Yeah **both are implemented using Lists datastructure** from python
+ But the secret lies in **accessing** the elements
+ if we access lists using **append,insert and remove** along with **+ve indices** for indexing elements it is **ARRAY**
+ if we access using **append for push and pop to remove** along with **-ve indices for indexing** elements which is LIFO it is **STACK**

```python
# Array implementation 

# Array (List) Implementation
array = [1, 2, 3, 4, 5]

# Access an element
print(array[2])  # Output: 3

# Append an element
array.append(6)  # Output: [1, 2, 3, 4, 5, 6]

# Insert an element at index 2
array.insert(2, 10)   # Output: [1, 2, 10, 3, 4, 5, 6]

# Remove an element by value
array.remove(4)

print(array)  # Output: [1, 2, 10, 3, 5, 6]



# Stack implementation
# Stack Implementation using list but this might cause confusion and errors if missed the flow of implementaiton 
stack = []

# Push elements to the stack
stack.append(1)
stack.append(2)
stack.append(3)

# Pop an element from the stack
top_element = stack.pop()
print(top_element)  # Output: 3

# Peek at the top element without removing it
top_element = stack[-1]  # Access the top element
print(top_element)  # Output: 2

# Check if the stack is empty
is_empty = len(stack) == 0
print(is_empty)  # Output: False


#So creating a class for stack implementation is effecient
class Stack:
    def __init__(self):
        self.stack = []

    def push(self, item):
        self.stack.append(item)

    def pop(self):
        if not self.is_empty():
            return self.stack.pop()
        return None

    def peek(self):
        if not self.is_empty():
            return self.stack[-1]
        return None

    def is_empty(self):
        return len(self.stack) == 0

# Using the Stack class
stack = Stack()
stack.push(1)
stack.push(2)
print(stack.peek())  # Output: 2
print(stack.pop())   # Output: 2
```

```python
'''
https://leetcode.com/problems/valid-parentheses/
https://www.youtube.com/watch?v=WTzjTskDFMg
Time complexity = O(n) as we are iterating through each element
Space complexity = O(n) as we use extra memory fo stack and hashmap  '''

class Solution:
    def isValid(self, s: str) -> bool:

        #creating an empty list
        stack = []

        # identify close to open brackets so if we insert close brackets when there is open bracket in stack it helps to pop those two
        CloseToOpen = { ")" : "(" , "]" : "[", "}" : "{" }

        
        for c in s:                    # taking 1st bracket from input
            if c in CloseToOpen:       # checking if that is key in closetoopen dict is not we append it to out list(stack)
                if stack and stack[-1] == CloseToOpen[c]:    # is list is not empty and its last element is equal to value of dict
                    stack.pop()                              # then we pop those brackets
                else:
                    return False
            else:
                stack.append(c)
        return True if not stack else False                  # finally returing true if list is empty



'''
https://leetcode.com/problems/min-stack/description/
https://www.youtube.com/watch?v=qkLl7nAwDPo
Time comp. = O(1) for everything for top it needs to fetch all stack so we are maintaining 
one more stack called minStack where it always stores the latest min value inside it 
'''

class MinStack:
    def __init__(self):      #here we are initializing 2 lists which records the value of stack and minstack
        self.stack = []
        self.minStack = []
        
    def push(self, val: int) -> None:
        self.stack.append(val) 
#takes min of given val & minstack last element  # if minstack is not empty it takes minstack data else takes val value
        val = min(val, self.minStack[-1] if self.minStack else val)  
        self.minStack.append(val) 
        
    def pop(self) -> None:
        self.stack.pop()
        self.minStack.pop()
        
    def top(self) -> int:
        return self.stack[-1]
        
    def getMin(self) -> int:
        return self.minStack[-1]
        

'''
https://leetcode.com/problems/baseball-game/submissions/1232020150/
https://www.youtube.com/watch?v=Id_tqGdsZQI
time complexity = O(n) as we go to each iteration 
'''
class Solution:
    def calPoints(self, operations: List[str]) -> int:
        stack = []
        for op in operations:
            if op == '+':
                stack.append(stack[-1] + stack[-2])
            elif op == 'D':
                stack.append(2 * stack[-1])
            elif op == 'C':
                stack.pop()
            else:
                stack.append(int(op))
        return sum(stack) 
```
