# Insertion Sort

- For an unsorted array or linkedlists, if there are previous element it checks the element with previous whether they are in sorted order or not. If not swaps both the elements in sorted order (ascending or descending)
- This happens until the indexed element is in sorted form as compared to their previous element
- Then it shifts to next indexed element and check whether it is in sorted form as compared to its previous one, if not swaps and sorts the elements. and again checks if they are in sorted form while comparing to its all previous indexed elements
- This goes on until it sortes all the elements

- Time complexity is __O(n^2)__ when it swaps the element for sorting and then again checking for all its previous elements for sorting
- If only swap happens and all the upcoming index elements are greater than the previous index elements time complexity will be __O(n)__ 

There are stable and unstable type of sorting
- __Stable Sorting__ = In above 7(brown),3(orange),7(yellow) example the sorting happens 3,7(brown),7(yellow) here 7 is common so no swap needed and 7(brown) is ordered first as it is first in 7(brown),3(orange),7(yellow) which is stable where unnecesssary swap doesn't happen
- __Unstable Sorting__ = But here despite being 7 are same 7(yellow) is swapped with 7(brown). So unnecessary swap happened which is unstable

```python
# Stable sorting with T- O(n^2), S- O(1)
'''
Implement Insertion Sort and return intermediate states.

Insertion Sort is a simple sorting algorithm that builds the sorted list one element at a time, from left to right.
It works by repeatedly taking an element from the unsorted portion and inserting it into its correct position in the sorted portion
of the list.

Objective:

Given a list of key-value pairs, sort the list by key using Insertion Sort.
Return a list of lists showing the state of the array after each insertion.
If two key-value pairs have the same key, maintain their relative order in the sorted list.

Input:
pairs - a list of key-value pairs, where each key-value has an integer key and a string value. (0 <= pairs.length <= 100).

Example 1:

Input:
pairs = [(5, "apple"), (2, "banana"), (9, "cherry")]

Output:
[[(5, "apple"), (2, "banana"), (9, "cherry")], 
 [(2, "banana"), (5, "apple"), (9, "cherry")], 
 [(2, "banana"), (5, "apple"), (9, "cherry")]]


Notice that the output shows the state of the array after each insertion.
The last state is the final sorted array. There should be pairs.length states in total.

Example 2:
Input:
pairs = [(3, "cat"), (3, "bird"), (2, "dog")]

Output:
[[(3, "cat"), (3, "bird"), (2, "dog")], 
 [(3, "cat"), (3, "bird"), (2, "dog")],
 [(2, "dog"), (3, "cat"), (3, "bird")]]

In this example, you can observe that the pairs with key=3 ("cat" and "bird") maintain their relative order,
illustrating the stability of the Insertion Sort algorithm.
'''

# Definition for a pair.
# class Pair:
#     def __init__(self, key: int, value: str):
#         self.key = key
#         self.value = value
class Solution:
    def insertionSort(self, pairs: List[Pair]) -> List[List[Pair]]:
        n = len(pairs)
        res = []        #intermediate result

        for i in range(n):
            j = i - 1 
            while j >=0 and pairs[j].key > pairs[j + 1].key:
                pairs[j], pairs[j+1] = pairs[j+1], pairs[j]      #swaping the values
                j -= 1                                           # this helps in checking all the previous indexed values are sorted w.r.t 
                                                                    # present index value or not 
            res.append(pairs[:])        # copy of pairs is adding to res
        return res
```
