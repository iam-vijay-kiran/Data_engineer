# Prefix SUM 




- Prefix Sum = sum of continuous values of an subarray is called prefix sum & it can be multiplication as well
- if an array is given if we always want to calculate the sum of subarray then it becomes linear time complexity i.e,O(N)
- so we impliment a data structure where all the elements are added linearly and placed as shown in green colored sum value 




- for calulating a sub array we take their range like L & R, here it is -3 & -3 respectively as shown above with blue colored. 
- Then we take R place sum value i.e., 1 and subtract its value from L-1 sum value i.e., 1 which are in green colored here.



- if the L-1 is out of bound i.e., there is no sum value we take 0 as default value and subtract from R position sum value to get sum of that sub array. we do impliment the same for multiplication too

- initially for adding all the array values it might take linar __time complexity O(N)__
- Because we do this and store in prefix array, later whenever we take sum of substring it takes constant __time complexity O(1)__
- initially if we dont maintain a prefix array and always add substrings whenever needed it always takes O(N) for every substring sum which is not effecient

- Prefix Sum = operations done from left to right

- Postfix Sum = Same as Prefix Sum but Right to Left Operations

```python
''' https://leetcode.com/problems/range-sum-query-immutable/description/
    https://www.youtube.com/watch?v=2pndAmo_sMA
    Initially it might take O(N) for computing sum of entire array values into self.prefix
    later whenever we call sum for substrings/array it takes only  O(1) time complexity
    S - O(n)
'''

'''
Input
["NumArray", "sumRange", "sumRange", "sumRange"]
[[[-2, 0, 3, -5, 2, -1]], [0, 2], [2, 5], [0, 5]]
Output
[null, 1, -1, -3]

'''
class NumArray:
    def __init__(self, nums: List[int]):
        self.prefix=[]          # creating an array where we append cumilated sum of elements
        curr = 0                # initializing the sum as 0
        for i in nums:          # looping through the array
            curr += i           # sum
            self.prefix.append(curr)     # append the value to new array

    def sumRange(self, left: int, right: int) -> int:
        prefixRight = self.prefix[right]                         # take the right range of element to calculate
        prefixLeft = self.prefix[left-1] if left > 0 else 0      # take left - 1 value if left > 0 else 0 
        return prefixRight - prefixLeft                         # sub left sum value from right value we get value for range of that subarray


# Your NumArray object will be instantiated and called as such:
# obj = NumArray(nums)
# param_1 = obj.sumRange(left,right)



# Brute force which takes T - O(n) for sub array calculation, S - O(1)
class NumArray:
    def __init__(self, nums):
        self.nums = nums

    def sumRange(self, left, right):
        res = 0
        for i in range(left, right + 1):
            res += self.nums[i]
        return res
```

```python
'''
https://www.youtube.com/watch?v=KE8MQuwE2yA
https://leetcode.com/problems/range-sum-query-2d-immutable/description/
'''

# Brute force which is T - O(n^2) as searching for each row and column. But S - O(1)
class NumMatrix:

    def __init__(self, matrix: list[list[int]]):
        self.matrix = matrix

    def sumRegion(self, row1: int, col1: int, row2: int, col2: int) -> int:
        res = 0
        for r in range(row1, row2 + 1):
            for c in range(col1, col2 + 1):
                res += self.matrix[r][c]
        return res


# Lest make an optimal way to solve the problem using 2-Dimensional Prefix Sum
# T - O(n) for initial sum of all elements but O(1) for particular range of elements
# S - O(1), as we use some extra space to store sum of calculations done







'''
https://leetcode.com/problems/find-pivot-index/
https://www.youtube.com/watch?v=u89i60lYx8U
'''
# prefix sum optimal, T- O(n) as count the nums linearly, S- O(1)
class Solution:
    def pivotIndex(self, nums: List[int]) -> int:
        total = sum(nums)               # take nums sum
        leftSum = 0                     # init. leftsum
        for i in range(len(nums)):      # iterating i to find pivot index    
            rightSum = total - nums[i] - leftSum        # rightsum = total - that i value - leftsum
            if leftSum == rightSum:         # if they are equal 
                return i                    # return pivot index
            leftSum += nums[i]              # else add that i value to leftsum
        return -1                       # if haven't found return -1



#Brute force T-O(n^2), S-O(1)
class Solution:
    def pivotIndex(self, nums: List[int]) -> int:
        n = len(nums)
        for i in range(n):
            leftSum = rightSum = 0
            for l in range(i):
                leftSum += nums[l]
            for r in range(i + 1, n):
                rightSum += nums[r]
            if leftSum == rightSum:
                return i
        return -1
```
