

- we take L, R pointers at starting and ending of string respectively. then compare them if same there are palindrome else not.
- Then increment L, decrement R and repeat the process until L < R to avoid duplicate comparision & out of bounds
- T = O(n), S = O(1) (no extra space is used)

```python
'''
https://leetcode.com/problems/valid-palindrome/
https://www.youtube.com/watch?v=jJXJ16kPFWg
'''
# using inbuit function, T = S = O(n)
class Solution:
    def isPalindrome(self, s: str) -> bool:
        newStr = ""                     # init. new string which is extra space using

        for char in s:                  
            if char.isalnum():          # validating char as alpha numeric using inbuilt function
                newStr += char.lower()  # adding to new String
        return newStr == newStr[::-1]   # compare string with reverse string gives boolean output

# without inbuilt function, T = O(n), S = O(1) as we take no extra space
class Solution:
    def isPalindrome(self, s: str) -> bool:
        l, r = 0 , len(s) - 1       # init. starting & ending pointers

        while l < r:                # we consider only when l pointer < r pointer to avoid duplicate processing
            while l < r and not self.isalnum(s[l]):     # if not alpha num increment l pointer
                l += 1
            while r > l and not self.isalnum(s[r]):     # if not alpha num decrement r pointer
                r -= 1
            if s[l].lower() != s[r].lower():          # convert to lower case and if not equal return false else true
                return False
            l , r = l+1, r-1        # moving to next elements of l,r pointers
        return True


    def isalnum(self, c):               # helper function checks c is in b/w respective required ascii values
        return (ord('A') <= ord(c) <= ord('Z') or
                ord('a') <= ord(c) <= ord('z') or
                ord('0') <= ord(c) <= ord('9'))




'''
https://leetcode.com/problems/two-sum-ii-input-array-is-sorted/
https://www.youtube.com/watch?v=cQ1Oz4ckceM
'''
# 2 pointers approach, T - O(n) and S - O(1) as no extra space is utilized & optimal solution
class Solution:
    def twoSum(self, numbers: List[int], target: int) -> List[int]:
        l, r = 0, len(numbers) - 1      # init l, r pointers

        while l < r :                   # avoiding duplicate process
            curSum = numbers[l] + numbers[r]    #take the sum
            if curSum > target:         
                r -= 1
            elif curSum < target:
                l += 1
            else:
                return [l+1, r+1]       # returning +1 cuz the indexing started from 1 not 0 as per problem statement
        return []

# Brute force T- O(n^2) and S- O(1)
class Solution:
    def twoSum(self, numbers: List[int], target: int) -> List[int]:
        for i in range(len(numbers)):
            for j in range(i + 1, len(numbers)):
                if numbers[i] + numbers[j] == target:
                    return [i + 1, j + 1]
        return []





'''
https://leetcode.com/problems/remove-duplicates-from-sorted-array-ii/
https://www.youtube.com/watch?v=ycAq8iqh0TI
'''
# 2 pointer approach & effecient one   T - O(n), S- O(1)
class Solution:
    def maxArea(self, height: List[int]) -> int:
        res = 0
        l, r = 0, len(height) - 1       # init 2 pointers 

        while l < r:                # avoid duplicate processing while l,r increment,decrement respectively
            area = (r - l) * min(height[l], height[r])      # length of 2 pointers * height of min(2 pointers)
            res = max(res, area)        # we want to the max value

            if height[l] < height[r]:       # checking whether still there are any other in entire list
                l += 1
            elif height[l] > height[r]:  
                r -= 1
            else:
                r -= 1              # as ifelse and this same we can combine both into else condition
                # or l += 1
        return res

# Brute Force, T - O(n^2), S - O(1)
class Solution:
    def maxArea(self, heights: List[int]) -> int:
        res = 0
        for i in range(len(heights)):
            for j in range(i + 1, len(heights)):
                res = max(res, min(heights[i], heights[j]) * (j - i))
        return res
```
