# Arrays and their time complexity

### 1. Collection of similar datatypes at consicutive memory location is called an array
### 2. Python has only Dynamic arrays which means the size of array is not pre-defined and flexible
+ Array with fixed memory and cannot add more elements without adjusting the existing elements
- Static arrays are used when memory is fixed and prior known
- Dynamic arrays are used when size of memory is unknown prior and are defalut in python & javascript
- Ambratization - doubling the array size to increase space capacity which is **O(N)** inserting there is **O(1)**

```python
''' Remove duplicates = https://leetcode.com/problems/remove-duplicates-from-sorted-array/submissions/1231943121/
    Time complexity = O(n) as it moves through each element
   Space complexity = O(1) as it make operations in existing array itself '''
list=[1,2,2,3,4,4,5]
class Solution:
    def removeDuplicates(self, nums: List[int]) -> int:
        l = 1
        for r in range(1,len(nums)):
            if nums[r] != nums[r-1]:
                nums[l] = nums[r]
                l +=1
        return l

'''        
 we initiliaze l,r at 1 not 0 because whenever we update the 0th value it stays in same 
 position so, we are checking from 1st position.      
 taking l as left pointer and comparing with r as right pointer if they are not equal 
 placing the element in left pointer 
'''


'''
 https://leetcode.com/problems/remove-element/
 removes the val element from list
 Time complexity = O(n) as it moves through each element'''
nums = [0,1,2,2,3,0,4,2], val = 2
class Solution:
    def removeElement(self, nums: List[int], val: int) -> int:
        k = 0
        for i in range(len(nums)):
            if nums[i] != val:
                nums[k] = nums[i]
                k += 1
        return k

    
'''
 https://leetcode.com/problems/concatenation-of-array/submissions/1232002497/
 https://www.youtube.com/watch?v=68isPRHgcFQ
 Time complexity = O(n) as we iterate through each element for both below codes
'''
class Solution:
    def getConcatenation(self, nums: List[int]) -> List[int]:
        return nums * 2
         
          #or 
        
class Solution:
    def getConcatenation(self, nums: List[int]) -> List[int]:
        ans = []
        for i in range(2):
            for n in nums:
                ans.append(n)
        return ans
```
